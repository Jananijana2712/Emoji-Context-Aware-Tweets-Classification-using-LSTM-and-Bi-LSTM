```python
import matplotlib.pyplot as plt
import numpy as np

# Data for precision, recall, and f1-score for LSTM and Bi-LSTM
metrics = ["Precision", "Recall", "F1-Score"]
classes = ["Class 0", "Class 1"]

# Top 10 Results
lstm_top10 = {"precision": [1.00, 0.94], "recall": [0.04, 1.00], "f1-score": [0.07, 0.97]}
bilstm_top10 = {"precision": [1.00, 0.94], "recall": [0.04, 1.00], "f1-score": [0.07, 0.97]}

# Top 5000 Results
lstm_top5000 = {"precision": [1.00, 0.99], "recall": [0.79, 1.00], "f1-score": [0.88, 0.99]}
bilstm_top5000 = {"precision": [1.00, 0.99], "recall": [0.79, 1.00], "f1-score": [0.88, 0.99]}

# Plotting function
def plot_comparison(metric, lstm_values, bilstm_values, dataset, metric_name):
    x = np.arange(len(classes))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, lstm_values, width, label='LSTM')
    rects2 = ax.bar(x + width/2, bilstm_values, width, label='Bi-LSTM')

    # Add some text for labels, title, and custom x-axis tick labels
    ax.set_ylabel(metric_name)
    ax.set_title(f'{metric_name} Comparison for {dataset}')
    ax.set_xticks(x)
    ax.set_xticklabels(classes)
    ax.legend()

    # Annotating the bars
    for rect in rects1 + rects2:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

    plt.ylim(0, 1.1)  # Set y-axis limit
    plt.show()

# Generate plots
datasets = ['Top 10', 'Top 5000']
metrics_names = ['Precision', 'Recall', 'F1-Score']

# Plot for Top 10
for i, metric in enumerate(metrics):
    plot_comparison(metric, lstm_top10[metric.lower()], bilstm_top10[metric.lower()], 'Top 10', metrics_names[i])

# Plot for Top 5000
for i, metric in enumerate(metrics):
    plot_comparison(metric, lstm_top5000[metric.lower()], bilstm_top5000[metric.lower()], 'Top 5000', metrics_names[i])
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, LSTM, Bidirectional, Dense, Dropout, Attention, Layer
from tensorflow.keras.models import Model

# Custom Attention Mechanism
class AttentionLayer(Layer):
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)

    def call(self, query, key, value):
        # Compute attention scores
        scores = tf.matmul(query, key, transpose_b=True)
        scores = tf.nn.softmax(scores, axis=-1)
        # Weighted sum of values
        output = tf.matmul(scores, value)
        return output

# Base function for LSTM and Bi-LSTM
def create_lstm_model(input_shape, units=64, bi_directional=False, attention=False):
    inputs = Input(shape=input_shape)
    if bi_directional:
        lstm_layer = Bidirectional(LSTM(units, return_sequences=True))
    else:
        lstm_layer = LSTM(units, return_sequences=True)
    
    x = lstm_layer(inputs)
    x = Dropout(0.3)(x)

    if attention:
        query = Dense(units)(x)
        key = Dense(units)(x)
        value = Dense(units)(x)
        attention_output = AttentionLayer()(query, key, value)
        x = attention_output

    x = LSTM(units)(x) if not bi_directional else Bidirectional(LSTM(units))(x)
    x = Dropout(0.3)(x)
    outputs = Dense(56, activation="softmax")(x)

    model = Model(inputs, outputs)
    return model

# Define input shape
input_shape = (30, 8)

# LSTM model
lstm_model = create_lstm_model(input_shape)
lstm_model.summary()

# Bi-LSTM model
bi_lstm_model = create_lstm_model(input_shape, bi_directional=True)
bi_lstm_model.summary()

# LSTM + Attention model
lstm_attention_model = create_lstm_model(input_shape, attention=True)
lstm_attention_model.summary()

# Bi-LSTM + Attention model
bi_lstm_attention_model = create_lstm_model(input_shape, bi_directional=True, attention=True)
bi_lstm_attention_model.summary()
```
