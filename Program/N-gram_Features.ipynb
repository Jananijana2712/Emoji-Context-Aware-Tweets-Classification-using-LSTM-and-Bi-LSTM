```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Load the dataset
file_path = "Processed_Tweets_Emoji.csv"  # Replace with your dataset's path
tweets_df = pd.read_csv(file_path)
```

```python
# Step 1: Preprocess to handle NaN values and ensure all entries are strings
tweets_df['Cleaned_Tweet_Emojified'] = tweets_df['Cleaned_Tweet_Emojified'].fillna("")  # Replace NaN with an empty string
tweets_df['Cleaned_Tweet_Emojified'] = tweets_df['Cleaned_Tweet_Emojified'].astype(str)  # Convert all entries to strings
```

```python
# Define a function for N-gram extraction
def extract_ngrams(tweets, ngram_range=(1, 3), max_features=10000):
    """
    Extracts N-grams (unigrams, bigrams, trigrams, etc.) from a text corpus.

    Args:
        tweets (pd.Series): A series containing the text data (tweets).
        ngram_range (tuple): The range of n-grams to extract (e.g., (1, 3) for unigrams, bigrams, and trigrams).
        max_features (int): Maximum number of features to keep.

    Returns:
        pd.DataFrame: DataFrame containing N-gram features for each text.
    """
    vectorizer = CountVectorizer(ngram_range=ngram_range, max_features=max_features)
    ngram_matrix = vectorizer.fit_transform(tweets)

    # Convert to a DataFrame
    ngram_df = pd.DataFrame.sparse.from_spmatrix(
        ngram_matrix, 
        columns=vectorizer.get_feature_names_out()
    )
    return ngram_df
```

```python
# Step 2: Apply N-gram extraction
ngram_range = (1, 3)  # Unigrams, bigrams, and trigrams
ngram_features = extract_ngrams(tweets_df['Cleaned_Tweet_Emojified'], ngram_range=ngram_range)
```

```python
# Step 3: Concatenate the N-gram features with the original DataFrame
tweets_with_ngrams = pd.concat([tweets_df.reset_index(drop=True), ngram_features.reset_index(drop=True)], axis=1)
```
